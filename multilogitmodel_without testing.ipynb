```python
import biogeme as db
import biogeme.database as db
import biogeme.biogeme as bio
import biogeme.models as models
import biogeme.version as ver
from biogeme.expressions import Beta
df=pd.read_csv('/content/datasetV1.xls - intcity.csv')
df['car1'] = df['car'] + df['air'] + df['rail']
df['rail1'] = df['car'] + df['air'] + df['rail']
df['air1'] = df['car'] + df['air'] + df['rail']
database=db.Database("testmodel",df)
globals().update(database.variables)

#directly access the names of variables in data file
#from headers import*

#list of parameters
ASC_C = Beta('ASC_car',0,None,None,0)
ASC_A = Beta('ASC_air',0,None,None,0)
ASC_R = Beta('ASC_rail',0,None,None,1)

#Generic Coefficients
BETA_IVTT = Beta('BETA_IVTT',0,None,None,0)
BETA_OVTT = Beta('BETA_OVTT',0,None,None,0)
BETA_COST = Beta('BETA_COST',0,None,None,1)


BETA_INCOME_c = Beta('BETA_INCOME_c',0,None,None,0)
BETA_INCOME_a = Beta('BETA_INCOME_a',0,None,None,0)
BETA_INCOME_r = Beta('BETA_INCOME_r',0,None,None,1)




#definition of utility function

U1= ASC_C+\
    BETA_IVTT*IVTT_c+\
    BETA_OVTT*OVTT_c+\
    BETA_INCOME_c*INCOME+\
    BETA_COST*COST_c

U2= ASC_A+\
    BETA_IVTT*IVTT_a+\
    BETA_OVTT*OVTT_a+\
    BETA_COST*COST_a+\
    BETA_INCOME_a*INCOME
U3= ASC_R+\
    BETA_IVTT*IVTT_r+\
    BETA_OVTT*OVTT_r+\
    BETA_COST*COST_r+\
    BETA_INCOME_r*INCOME


#Associate the Utility with some numbers
U={1:U1,2:U2,3:U3}

#Availability of each alternative as described in the dataset
AV = {1:car1,2:air1,3:rail1}


#choice model is a logit model with availability conditions
logprob=models.loglogit(U,AV,CHALT)

biogeme=bio.BIOGEME(database,logprob)
biogeme.modelName ='testmodel'
results=biogeme.estimate()
pandasresults=results.getEstimatedParameters()
pandasresults

```
